```python
import pickle
import matplotlib.pyplot as plt
```

```python
bart_trt_times = pickle.load(open('/data/yuchen/projects/gns/data/bart_trt_times.pkl','br'))
bart_raw_times = pickle.load(open('/data/yuchen/projects/gns/data/bart_raw_times.pkl','br'))
bart_trt_time=[e['raw_time'] for e in bart_trt_times]
bart_raw_time=[e['raw_time'] for e in bart_raw_times]
ax2= plt.subplot(111)
ax2.plot(bart_trt_time,label='bart_trt_time')
ax2.plot(bart_raw_time,label='bart_raw_time')
legend = ax2.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```

```python
gen_trt_times = pickle.load(open('/data/yuchen/projects/gns/data/gen_trt_times.pkl','br'))
gen_raw_times = pickle.load(open('/data/yuchen/projects/gns/data/gen_raw_times.pkl','br'))
trt1=[e[0] for e in gen_trt_times]
raw1=[e[0] for e in gen_raw_times]
trt2=[e[1] for e in gen_trt_times]
raw2=[e[1] for e in gen_raw_times]
trt3=[e[2] for e in gen_trt_times]
raw3=[e[2] for e in gen_raw_times]
trt4=[e[3] for e in gen_trt_times]
raw4=[e[3] for e in gen_raw_times]
trt5=[e[4] for e in gen_trt_times]
raw5=[e[4] for e in gen_raw_times]
trt6=[e[5] for e in gen_trt_times]
raw6=[e[5] for e in gen_raw_times]
ax2= plt.subplot(111)
ax2.plot(trt1,label='trt1')
ax2.plot(raw1,label='raw1')
ax2.plot(trt2,label='trt2')
ax2.plot(raw2,label='raw2')
ax2.plot(trt3,label='trt3')
ax2.plot(raw3,label='raw3')
ax2.plot(trt4,label='trt4')
ax2.plot(raw4,label='raw4')
ax2.plot(trt5,label='trt5')
ax2.plot(raw5,label='raw5')
ax2.plot(trt6,label='trt6')
ax2.plot(raw6,label='raw6')
ax2.plot(trt1+trt2+trt3+trt4+trt5+trt6,label='trt_all')
ax2.plot(raw1+raw2+raw3+raw4+raw5+raw6,label='raw_al')
legend = ax2.legend(loc=0, shadow=True,bbox_to_anchor=(1, 1), fontsize='x-large')
plt.show()
```
